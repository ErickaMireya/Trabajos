```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plp

F=np.array([-5,1,9,25,55,105])
X=np.array([0,1,2,3,4,5])

plt.scatter(X,F,color="orange",marker="*")
plt.title("Grafica de los datos")
plt.xlabel ("Eje X")
plt.ylabel ("Eje y")
plt.grid(color="pink",linestyle="--")
plt.show()
```

```python
def f(x1):
    return x1**3-2*x1**2+7*x1-5
```
